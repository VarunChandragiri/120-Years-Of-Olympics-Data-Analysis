```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
Dataset1 = pd.read_csv('./DATA/athelete_events_1992_2016.csv')
```

```python
Dataset2 = pd.read_csv('./DATA/athlete_events_1896_1988.csv')
```

```python
Dataset3 = pd.read_csv('./DATA/noc_regions.csv')
```

```python
Dataset1.head()
```

```python
Dataset2.head()
```

```python
Dataset3.head()
```

```python
Dataset1.info()
```

```python
frames = [Dataset1,Dataset2]
result = pd.concat(frames)
```

```python
result.info()
```

```python
FinalData = pd.merge(result,Dataset3,on='NOC')
FinalData.head()
```

## Winners Data

```python
Winners = FinalData[FinalData['Medal'].notnull()]
Winners
```

### Top 10 cities hosted olympics

```python
Top_10_cities_hosted_olympics = FinalData['City'].value_counts().head(10)

Top_10_cities_hosted_olympics
```

```python
plt.barh(Top_10_cities_hosted_olympics.index,Top_10_cities_hosted_olympics)
plt.xlabel('No. of Medals')
plt.ylabel('Countries / Regions')
plt.title('Top 10 cities hosted olympics')
plt.show()
```

### Type of Medals and No. of medals given respectively

```python
FinalData['Medal'].value_counts()
```

```python
plt.scatter(FinalData['Medal'].value_counts().index,FinalData['Medal'].value_counts(), color='green')
plt.show()
```

### Top 10 countries with highest medals

```python
Top_10_countries_with_highest_medals = Winners.groupby('region').count().sort_values('Medal',ascending=False)['Medal'].head(10)

Top_10_countries_with_highest_medals
```

```python
plt.barh(Top_10_countries_with_highest_medals.index,Top_10_countries_with_highest_medals)
plt.xlabel('No. of Medals')
plt.ylabel('Countries / Regions')
plt.title('Top 10 countries with highest no. of medals')
plt.show()
```

### Top 10 countries with highest medals in single year

```python
Top_10_countries_with_highest_medals_in_single_year  = Winners.groupby(['Year','region']).count().sort_values('Medal',ascending=False)['Medal'].head(10)

Top_10_countries_with_highest_medals_in_single_year
```
